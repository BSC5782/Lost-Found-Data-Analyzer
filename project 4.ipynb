```python
import pandas as pd
df = pd.read_csv("lost_found.csv")
print(df)
```

```python
import pandas as pd
df["Date_Lost"] = pd.to_datetime(df["Date_Lost"])
df["Month"] = df["Date_Lost"].dt.month
```

```python
import pandas as pd 
print(df["Item_Name"].value_counts())
```

```python
import pandas as pd 
print(df["Location"].value_counts())
```

```python
import pandas as pd
print(df["Month"].value_counts().sort_index())
```

```python
import pandas as pd 
print(df["Status"].value_counts())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df["Item_Name"].value_counts().plot(kind="bar")
plt.title("Most Lost Items")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df["Location"].value_counts().plot(kind="bar")
plt.title("Loss by Location")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df["Month"].value_counts().sort_index().plot(kind="line", marker='o')
plt.title("Monthly Trend")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df["Status"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Found vs Unclaimed")
plt.ylabel("")
plt.show()
```
